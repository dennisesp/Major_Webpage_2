```python
selected_races = ['Hispanic or Latino']
selected_majors = ['26.01 - Biology, General.']


filtered_df = df_combined[
    (df_combined['Race and Ethnicity'].isin(selected_races)) &
    (df_combined['Major'].isin(selected_majors))
]


grouped_df = filtered_df.groupby(
    ['Race and Ethnicity',]
)['Awards/Degrees Conferred'].sum().reset_index()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style='whitegrid', context='talk')


latino_df = df_combined[df_combined['Race and Ethnicity'] == 'Hispanic or Latino']


major_counts = latino_df.groupby('Major')['Awards/Degrees Conferred'].sum().reset_index()


major_counts = major_counts.sort_values('Awards/Degrees Conferred', ascending=False)


major_counts['Major Short'] = major_counts['Major'].str.replace(r'^\d{2}\.\d{2,4} - ', '', regex=True)



plt.figure(figsize=(12, 8))
barplot = sns.barplot(
    data=major_counts,
    y='Major Short',
    x='Awards/Degrees Conferred',
    palette='crest'
)


plt.title('Top 20 Majors by Number of Latino Doctorate Graduates (2021–2023)', fontsize=16)
plt.xlabel('Number of Graduates')
plt.ylabel('Major')
plt.tight_layout()


for index, row in major_counts.iterrows():
    barplot.text(row['Awards/Degrees Conferred'], index, row['Awards/Degrees Conferred'], color='black', va="center")

plt.show()
```
